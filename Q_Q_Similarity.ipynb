###Using TfiDf question- Qestion similarity

```python
df_part1['tokenized_title2'] = df_part1['tokenized_title'].apply(lambda x: ' '.join(x))
df_part2['tokenized_title2'] = df_part2['tokenized_title'].apply(lambda x: ' '.join(x))

# combining the sentences from tow dataframes
all_sentences = pd.concat([df_part1['tokenized_title2'], df_part2['tokenized_title2']])

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(all_sentences)

tfidf_matrix_part1 = tfidf_matrix[:len(df_part1)]
tfidf_matrix_part2 = tfidf_matrix[len(df_part1):]

#for each sentence in df_part2
top_similar_sentences = []
top_similarity_scores = []
top_similar_sentences_INDEX = []

# top 5 similar sentences in df_part1 for each sentence in df_part2
for i in range(len(df_part2)):
    query_sentence_vector = tfidf_matrix_part2[i]
    similarity_scores = cosine_similarity(query_sentence_vector, tfidf_matrix_part1)

    top_indices = similarity_scores.argsort()[0][-5:][::-1] # index of the top 5 most similar sentence

    similar_sentences = df_part1.iloc[top_indices]['title2'].tolist() # actual sentences using the indices
    similar_sentences_INDEX = df_part1.iloc[top_indices]['IndexNumber'].tolist()
    similarity_scores = similarity_scores[0][top_indices].tolist()

    #the list for each row in df_part2
    top_similar_sentences.append(similar_sentences)
    top_similar_sentences_INDEX.append(similar_sentences_INDEX)
    top_similarity_scores.append(similarity_scores)

for j in range(5):
    df_part2[f'top_similar_sentence_{j+1}'] = [similar_sentences[j] for similar_sentences in top_similar_sentences]
    df_part2[f'top_similar_sentence_INDEX_{j+1}'] = [similar_sentences_INDEX[j] for similar_sentences_INDEX in top_similar_sentences_INDEX]
    df_part2[f'top_similar_score_{j+1}'] = [similarity_scores[j] for similarity_scores in top_similarity_scores]


selected_columns = ['title'] + [f'top_similar_sentence_{i+1}' for i in range(5)] + [f'top_similar_score_{i+1}' for i in range(5)] + [f'top_similar_sentence_INDEX_{i+1}' for i in range(5)]

df_selected = df_part2[selected_columns]

df_selected.to_excel("output_TFiDf_similarity.xlsx", index=False, encoding='utf-8')
df_selected
```

```python
from IPython import display
import requests
import time
import torch
from sentence_transformers import models, SentenceTransformer, util
```

```python
def load_st_model(model_name_or_path):
    word_embedding_model = models.Transformer(model_name_or_path)
    pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension(),
                                   pooling_mode_mean_tokens=True,
                                   pooling_mode_cls_token=False,
                                   pooling_mode_max_tokens=False)

    model = SentenceTransformer(modules=[word_embedding_model, pooling_model])
    print(model)
    return model
```

```python
dfc_sorted = dfc.sort_values(by='IndexNumber')

eighty_percent_index = int(0.8 * len(dfc_sorted))

corpus = dfc_sorted.iloc[:eighty_percent_index]
queries = dfc_sorted.iloc[eighty_percent_index:]

corpus = corpus['title2'].tolist()
queries = queries['title2'].tolist()

top_k = 5
```

```python
embedder = load_st_model('HooshvareLab/bert-base-parsbert-uncased')
corpus_embeddings = embedder.encode(corpus, convert_to_tensor=True, show_progress_bar=True)


results = []
for query in queries:
    query_embedding = embedder.encode(query, convert_to_tensor=True, show_progress_bar=True)
    cos_scores = util.pytorch_cos_sim(query_embedding, corpus_embeddings)[0]
    cos_scores = cos_scores.cpu()

    #torch.topk to find the highest 5 scores
    top_results = torch.topk(cos_scores, k=top_k)
    #dic for  query and its top k similars
    query_results = {}
    query_results['query'] = query

    for i, (score, idx) in enumerate(zip(top_results[0], top_results[1])):

        query_results[f'similar_sentence_{i+1}'] = corpus[idx]
        query_results[f'score_{i+1}'] = score.item()
        query_results[f'index_{i+1}'] = idx.item()  # we need the index number too



    results.append(query_results)

df_parsBERT= pd.DataFrame(results)
df_parsBERT.to_excel("output_bert_parsBERT_similarity.xlsx", index=False, encoding='utf-8')
```

```python
import csv
#Converting CSV file to Qrel format : Annotated_index_qreal_file

def convert_csv_to_qrel(csv_file, qrel_file):
    qrel = {}
    with open(csv_file, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        next(csvreader)  #Skip hedears of   the file
        for row in csvreader:
            query_Id, _, object_id, relevance = row
            if query_Id not in qrel:
                qrel[query_Id] = {}
            qrel[query_Id][object_id] = int(relevance)

    with open(qrel_file, 'w') as qrelfile:
        for query_Id, docs in qrel.items():
            for object_id, relevance in docs.items():
                qrelfile.write(f"{query_Id} {object_id} {relevance}\n")

csv_file = 'Annotated_index_qreal_file.csv'
qrel_file = 'Annotated_index_qreal_file.qrel'

convert_csv_to_qrel(csv_file, qrel_file)
```

```python
import pytrec_eval
# Read the qrel file and convert it to a dictionary format that pytrec_eval understands it

def read_qrel(qrel_file):
    qrel = {}
    with open(qrel_file, 'r') as f:
        for line in f:
            query_Id , object_id, relevance = line.strip().split(' ')
            if query_Id not in qrel:
                qrel[query_Id] = {}
            qrel[query_Id][object_id] = int(relevance)
    return qrel

qrel_file = 'Annotated_index_qreal_file.qrel'
qrel = read_qrel(qrel_file)
print(qrel)
```

```python
def convert_excel_to_trec_format(excel_file, output_file, method_name):
    data = pd.read_excel(excel_file)

    with open(output_file, 'w') as outfile:
        for index, row in data.iterrows():
            query_id = row['query_id']
            for i in range(1, 6):  # for finding 5 top
                doc_id = row[f'index_{i}']
                rank = i
                score = row[f'top_similar_score_{i}']
                outfile.write(f"{query_id} Q0 {doc_id} {rank} {score} {method_name}\n")

# excel_file = 'output_TFiDf_qrel_format.xlsx'
# output_file = 'output_TFiDf_trec_format.txt'
# method_name = 'Tfidf'

# convert_excel_to_trec_format(excel_file, output_file, method_name)
```

```python
def read_method_output(output_file):
    # read the method of TREC fromat output file and convert it to a dictionary format
    method_output = {}
    with open(output_file, 'r') as f:
        for line in f:
            query_id, _, doc_id, rank, score, method_name = line.strip().split()
            if query_id not in method_output:
                method_output[query_id] = []
            method_output[query_id].append({'doc_id': doc_id, 'rank': int(rank), 'score': float(score), 'method_name': method_name})
    return method_output
```

```python
# the result of our three models
models = ['TFiDf' , 'bert_fa3' , 'bert_parsBERT']

for model in models:
    excel_file = f'output_{model}_qrel_format.xlsx'
    output_file = f'output_{model}_trec_format.txt'
    method_name = model

    convert_excel_to_trec_format(excel_file, output_file, method_name)
    # output_file = 'output_TFiDf_trec_format.txt'
    output_file =  f'output_{model}_trec_format.txt'
    method_output = read_method_output(output_file)

    # change the format to the true format
    restructured_method_output = {}
    for query_id, docs in method_output.items():
        restructured_method_output[query_id] = {doc['doc_id']: doc['score'] for doc in docs}


    evaluator = pytrec_eval.RelevanceEvaluator(qrel, {'map', 'ndcg' , 'recip_rank'})
    results = evaluator.evaluate(restructured_method_output)

    total_map = 0.0
    total_ndcg = 0.0
    total_mrr = 0.0

    num_queries = len(results)

    for query_id, query_results in results.items():
        total_map += query_results['map']
        total_ndcg += query_results['ndcg']
        total_mrr += query_results['recip_rank']


    average_map = total_map / num_queries
    average_ndcg = total_ndcg / num_queries
    average_mrr = total_mrr / num_queries

    print(f"Model: {model}")
    print(f"Average MAP: {average_map}")
    print(f"Average nDCG: {average_ndcg}")
    print(f"Average MRR: {average_mrr}")
    print()
```
