```python
import csv
import json

# Open the JSON file and load the data
with open('hisalamat_urls.json', 'r') as json_file:
    data = json.load(json_file)

# Open a CSV file for writing
with open('data.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)

    # Write the header row
    writer.writerow(data[0])

    # Write the data rows
    for row in data:
        writer.writerow(row)
```

```python
df_f = pd.read_csv('data.csv')
df_f
```

```python
search_string = 'http://www.hisalamat.com:80/ques/'
mask = df_f['original'].str.contains(search_string)
new_df = df_f[mask]
new_df
```

```python
base_ur= "http://web.archive.org/web/"

new_df['Merged_url'] = new_df.apply(lambda row: base_ur + str(row['endtimestamp']) + '/' + row['original'] , axis=1)
new_df
```

```python
import requests
from bs4 import BeautifulSoup

docs = []
for url in new_df['Merged_url']:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    title = soup.find("div", class_= 'quesP')
    body = soup.find('div', class_= 'ansP')
    group = soup.find('div' , class_ ='col-md-4 col-sm-4 col-xs-12')
    doctor = soup.find('div' , class_ ='page-header')
    doc = {"title": title.text,
           "body" : body.text ,
           "group" : group.text ,
           "doctor" : doctor.text }
    docs.append(doc)
```

```python
docs
```

```python
import json
with open("x.json", "w") as file:
  json.dump(docs, file)
```

```python
# create a new DataFrame with the downloaded data
new_df = pd.DataFrame({'Data': docs})

# save the new DataFrame to a CSV file
new_df.to_csv('output.csv', index=False)
```
